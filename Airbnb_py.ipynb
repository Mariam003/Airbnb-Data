```python
#removing $, "," and " " from the price and service fee column to check the statistical report of all columns
df['price'] = df['price'].str.replace('$', '', regex = False).str.replace(',', '', regex = False).astype(float)
df['service fee'] = df['service fee'].str.replace('$', '', regex = False).str.replace(',', '',).astype(float)
```

# Data Analysis

```python
# price distribution
sns.boxplot(data=df, x = 'price')
```

```python
sns.histplot(data=df, x = 'price')
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.xlabel('price $')
```

```python
df.info()
```

```python
#availability 365 distribution
plt.figure(figsize = (6, 3))
sns.histplot(data=df, x = 'availability 365', bins=100)
plt.xlim(0, 370)
plt.title('availability 365 distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# price distribution for neighbourhood groups
df.groupby(by = 'neighbourhood group')['price'].mean
```

```python
# neighbourhood group by price
plt.figure(figsize = (8, 5))
sns.barplot(data=df, x='neighbourhood group', y='price', hue = 'room type')
plt.title('neighbourhood group by price')
```

```python
# price distribution for neighbourhood groups
sns.countplot(data = df, y = 'neighbourhood group', order=df['neighbourhood group'].value_counts().index)
plt.title('number of listings by neighbourhood group')
plt.xlabel('Count')
```

```python
#room type distribution
plt.figure(figsize=(8, 5))
plt.title('room type distribution')
plt.ylabel('Frequemcy')
sns.countplot(data = df, x = 'room type')
```

```python
# relationship btw price and roomtype
sns.boxplot(data=df, x='room type', y = 'price')
plt.title('room type by price')
```

```python
# changes in number of reviews over time
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
reviews_over_time.plot(kind='line')
plt.title('number of reviews over time')
plt.xlabel('Date')
plt.ylabel('Number Of Reviews')
```

```python
# Geographical distribution
plt.figure(figsize = (10, 7))
sns.scatterplot(data=df, x= 'long', y= 'lat')
plt.title('Geographical Distribution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```
