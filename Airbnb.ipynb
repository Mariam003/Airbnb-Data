# Data Analysis

```python
# price distribution
sns.boxplot(data=df, x = 'price')
```

```python
# The boxplot shows that there are no outliers and the prices of most airbnbs range from around $ 300 to $ 900
```

```python
sns.histplot(data=df, x = 'price')
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.xlabel('price $')
```

```python
# The chart above shows the distribution of prices in the dataset
# The frequency axis shows the number of airbnbs that fall into each price range. 
# Most listings fall between the 2000 and 2500 
# The bins show that the price distribution seems a bit uniform and there is no specific price that is higher than others
```

```python
df.info()
```

```python
#availability 365 distribution
plt.figure(figsize = (6, 3))
sns.histplot(data=df, x = 'availability 365', bins=100)
plt.xlim(0, 365)
plt.title('availability 365 distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# The histogram shows the distribution of airbnbs that are available all year round
# The tallest bar at 0-25/30 days show that most airbnbs are available for a short period or are booked for most of the year
# The other tall bar ar 325/330-350 days show that some listings are available for booking for almost the entire year
# Fewer listings fall between 100-300 days
```

```python
# price distribution for neighbourhood groups
plt.figure(figsize = (8, 5))
sns.barplot(data=df, x='neighbourhood group', y='price')
plt.title('neighbourhood group by price')
```

```python
# The bar chart shows the price distibution in the neighbourhood
# Most of the prices are in the same range not lesser than $600
```

```python
# minimum nights vs room type
plt.figure(figsize = (8, 5))
sns.barplot(data=df, x ='room type', y = 'minimum nights')
plt.title('minimum nights vs room type')
```

```python
# The bar chart shows the minimum nights spent in each room types 
# Most people who pick the entire apartment spend a minimum of 9 to 10 nights
# Those who pick a shared room spend between 6 to 8 nights, likewise private room
# for the hotel room, people spend a minimum of 4 nights
```

```python
# neighbourhood group by price
plt.figure(figsize = (8, 5))
sns.barplot(data=df, x='neighbourhood group', y='price', hue = 'room type')
plt.title('neighbourhood group by price')
```

```python
# The bar chart shows the prices differences in the neighbourhood groups separated by room types
# Most of the priices are in the same range except for the hotel room in Queens neighbourhood which is lower than the other two neighbourhoods
```

```python
# price distribution for neighbourhood groups
sns.countplot(data = df, y = 'neighbourhood group', order=df['neighbourhood group'].value_counts().index)
plt.title('number of listings by neighbourhood group')
plt.xlabel('Frequency')
```

```python
# Manhattan and Brooklyn have the most listings
# Listing availability is heavily concentrated in Manhattan and Brooklyn, with other neighbourhoods having much less representation 
```

```python
# relationship btw price and roomtype
sns.barplot(data=df, x='room type', y = 'price')
plt.title('room type by price')
```

```python
# The bar chart shows that all the prices are in the same range with hotel room being a little expensive than others
```

```python
# Geographical distribution
plt.figure(figsize = (10, 7))
sns.scatterplot(data=df, x= 'long', y= 'lat')
plt.title('Geographical Distribution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```

```python
# This scatter plot shows the geographical distribution of the dataset based on Latitude and Longitude coordinates.
# The dense clusters of blue dots form the shape of New York City's boroughs. This visualizes the geographical 
# spread of listings, showing where they are concentrated (Manhattan, Brooklyn, Queens, Bronx, Staten Island) and confirming the dataset's location.
```
